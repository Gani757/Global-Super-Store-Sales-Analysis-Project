```python
#check for outliers
plt.figure(figsize=(10,5))
sns.boxplot(df[continuous])
plt.show()
```

```python
#handling outlier using IQR 
for i in continuous:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    ll=Q1-(1.5*IQR)
    ul=Q3+(1.5*IQR)
    outliers = df[(df[i] < ll) | (df[i] > ul)]
    print(f"Outliers in column '{i}':")
    print(outliers[i].reset_index(drop=True))
    df=df[(df[i] > ll) & (df[i] < ul)].reset_index(drop=True)   # removed outliers stored in df
```

```python
df["Date"]=pd.to_datetime(df["Date"],errors="coerce")   # converted date to datetime format

df["yearmonth"]=df["Date"].dt.to_period("M")


monthly_data=df.groupby("yearmonth")[["Sales","Profit"]].sum().reset_index()

# convert yearmonth to timestamp for plotting 
monthly_data["yearmonth"]=monthly_data["yearmonth"].dt.to_timestamp() 


# stacked line plot --monthly trend showing sales + profit

plt.figure(figsize=(10, 5))
plt.plot(monthly_data["yearmonth"], monthly_data["Sales"],marker="s",label="sales")
plt.plot(monthly_data["yearmonth"], monthly_data["Profit"],marker="s",label="profit")
plt.legend()
plt.title("Monthly sales")
plt.xlabel("month")
plt.ylabel("sales")
plt.xticks(rotation=45)
plt.show()
```

### Insight:
Focus on increasing profit efficiency during high-sales months (optimize discounts or reduce costs).

Consider targeted campaigns in Q4, as sales historically spike during that period.

Investigate mid-year performance dip to identify root causes and fix gaps.

# Overall Insights

1) Sales & Profit Trend:

Sales and Profit showed noticeable ***monthly fluctuations***, with peaks observed around ***Octâ€“Nov 2013 and Octâ€“Nov 2014.***

Despite sales rising, profit margins remained lower, indicating potential discount-driven or high-cost sales.

2) Top Performing Countries:

***United States, Mexico, and France*** generated the highest overall sales and profit.

However, the ***Government segment consistently outperformed*** others across countries in terms of profit.

3) Product Performance:

***Paseo*** emerged as the best-selling and ***most profitable product***, far ahead of others.

Products like Montana and Velo, although generating revenue, showed comparatively lower profitability, indicating opportunities for promotion or bundling.

4) Segment Insights:

The ***Government segment dominated*** both ***sales and profitability.

Channel Partners and Midmarket showed moderate performance; custom strategies may be needed for better conversion.

5) Discount & Profit Correlation:

A moderate ***negative correlation*** observed between Discount and Profit (â€“0.51) suggests ***excessive discounting*** could be ***hurting profitability***.

6) Category & Bundling Opportunities:

Products with high sales but lower margins (e.g., Carretera, Montana) are good candidates for bundling with top performers like Paseo to boost overall order value
